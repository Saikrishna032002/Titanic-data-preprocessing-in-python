```python



survival_counts = pd.crosstab(df['pclass'], df['survived'])

plt.figure(figsize=(6, 4))

ax = sns.heatmap(
    survival_counts, 
    annot=True,     
    fmt='d',        
    cmap='Blues',   
    cbar_kws={'label': 'Number of Passengers'}
)

# --- Customization ---
plt.title('Passenger Class vs. Survival Count (Heatmap)', fontsize=16, fontweight='bold')
plt.xlabel('Survived (0=No, 1=Yes)', fontsize=12)
plt.ylabel('Passenger Class', fontsize=12)


ax.set_yticklabels(['1st Class', '2nd Class', '3rd Class'], rotation=0)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(data= df, x='age', hue='survived', multiple='stack', palette='pastel')
plt.title('Survived vs Age (histplot)', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(7, 6))

ax = sns.barplot(data=df, x='embarked', y='survived', palette='tab10')

plt.title('Survival Rate by Embarkation Port (barplot)', fontsize=16, fontweight='bold')
plt.xlabel('Embarked (C = Cherbourg, Q = Queenstown, S = Southampton)', fontsize=12)
plt.ylabel('Survival Rate (Proportion of Survivors)', fontsize=12)

plt.ylim(0, 1)

plt.show()
```
